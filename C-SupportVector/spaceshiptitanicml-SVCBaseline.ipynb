# **Creation of the Pipeline**

Time for some column transformations and changes.

#### We do not really need **'PassengerId'** as it has no importance to our target value. We can safely dispose of it.

```python
train = train.drop(columns=['PassengerId'], axis=1)
```

```python
train.head()
```

#### **Lets start with some Column Transformations**

Here is how we are going to go:
* Transform the categorical columns with OHE
* Scale the numerical columns
* Apply model
* Test accuracy

##### **First, lets define the categorical features and numerical features:**

But first, lets look at the number of unique values in each of these categorical features

```python
categorical_features = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']

for i in categorical_features:
    print(f'The number of unique values in {i}: {train[i].nunique()}')
```

#### **Okay, looks like they are One Hot Encoding feasible.**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

categorical_transformer = OneHotEncoder(sparse_output = False, handle_unknown='ignore')
```

#### **Now, we will scale the numerical columns.**

Create a transformer for the numerical columns.

Standard scaling.

```python
from sklearn.preprocessing import StandardScaler

numerical_transformer = StandardScaler()
```

Now lets list all the numerical features we want to scale:

```python
train.head()
```

```python
numerical_features = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
```

### **Now we bundle them!**

```python
preprocessor = ColumnTransformer(transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num', numerical_transformer, numerical_features)])
```

# **Logistic Regression**

First, we will try out logistic regression and see how it works.

```python
from sklearn.linear_model import LogisticRegression

lgr_clf = LogisticRegression(random_state=1)
```

### **Lets create our first Pipeline for Logistic Regression**

Pipeline manual:
* The 'preprocessor' step basically does what we previously describe the 'preprocessor' variable to do i.e. Column Transformations
* The 'model' step fits the model onto the dataset after preprocessing.

```python
pipeline_lgr = Pipeline(steps=[('preprocessor', preprocessor),
                               ('model', lgr_clf)])
```

##### **Now we split the training data into X and Y, to test predictions.**

```python
X_train = train.iloc[:,0:10]
y_train = train.iloc[:, -1]
```

```python
X_train.head()
```

```python
pipeline_lgr.fit(X_train, y_train)
```

## **Imputation guide:**

* **Numerical Imputations:** Mean
* **Categorical Imputations :** Mode

```python
from sklearn.impute import SimpleImputer

for column in numerical_features:
    test[column] = test[column].fillna(test[column].mean())
    
test['HomePlanet']= test.HomePlanet.fillna('Earth')
test['CryoSleep']= test.CryoSleep.fillna('False')
test['Destination']= test.Destination.fillna('TRAPPIST-1e')
test['VIP']= test.VIP.fillna('False')
```

### **Here is what we want to go about this:**
* Make these different classifiers go through a simple loop


```python
from sklearn.ensemble import RandomForestClassifier
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score


pipeline_mult = Pipeline(steps=[('preprocessor', preprocessor)]) 

model_mult = pipeline_mult.fit(X_train, y_train)
```

```python
models = [RandomForestClassifier(n_estimators=100, random_state=1), GaussianNB(), SVC(random_state=1)]
model_name = ["Random Forest Classifier", "Naive Bayes", "SVM"]

for model, model_name in zip(models, model_name):
    pipeline_multi = Pipeline(steps=[('preprocessor', preprocessor),
                                   ('model', model)]) 
    pipeline_multi.fit(X_train, y_train)
    print(model_name)
    start = time.time()
    for score in ["accuracy", "precision", "recall"]:
        print(f'{score}: {cross_val_score(pipeline_multi, X_train, y_train, scoring=score, cv=5).mean()}')        
    print(f'{time.time() - start}\n')
```

## Grid Search for Random Forest Classifier 

Parameters to use:
* n_estimators (obviously)  - #Number of tress in the Random Forest
* max_features - #Number of features to consider at every split
* max_depth - #Maximum number of levels in the tree
* min_simples_split - #Minimum number of samples required to split a node
* min_samples_leaf - #Minimum number of samples required at each leaf node
* bootstrap - #Method of selecting samples for training each tree

```python
# Number of tress in the Random Forest
n_estimators = [int(x) for x in np.linspace(start=10, stop=80, num=10)]
# Number of features to consider at every split
max_features = ['sqrt']
# Maximum number of levels in the tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2,5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1,2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

Now, lets create the parameter grid:

```python
param_grid_rfc = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf,
              'bootstrap': bootstrap}
print(param_grid_rfc)
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

rfc_model = RandomForestClassifier()

from sklearn.model_selection import GridSearchCV

rfc_Grid = GridSearchCV(estimator = rfc_model, n_jobs=4, param_grid=param_grid_rfc, cv=10, verbose=2)

start = time.time()

pipeline_grid_rfc = Pipeline(steps=[('preprocessor', preprocessor),
                                ('gridsearch', rfc_Grid)])

pipeline_grid_rfc.fit(X_train, y_train)

print(f'\n{time.time() - start}')
```

### **Best Parameters:**

## **Extreme Gradient Booster Classifier**



**Steps**

* Creating a function for getting the accuracy for different n_estimator (Similar to what we did for Random Forest Classifier)
* Make it go through a for loop
* Visualize to see the best n_estimator for XGBClassifier

```python
from xgboost import XGBClassifier

def get_accuracy_xgb(n_estimators):
    local_pipeline_xgb = Pipeline(steps=[('preprocessor', preprocessor),
                                    ('model', XGBClassifier(n_estimators=n_estimators, random_state=1))])
    scores = cross_val_score(local_pipeline_xgb, X_train, y_train, cv=3, scoring='accuracy').mean()
    
    return scores


result1 = {}
for i in range(1,9):
    result1[50*i] = get_accuracy_xgb(50*1)  
    
sns.lineplot(x=list(result1.keys()), y=list(result1.values()))
plt.show()
```

**The model looks like it already reached its optimal n_estimators at n_estimators = 50, so we see a similar situation here as Random Forest Classifier**

### **Parameter Tuning/ Grid Search for XGBClassifier**

#### First, we create the **estimator** i.e. the **XGBClassifier** model.

In the **'objective' parameter**, I put it as **"binary:logistic"** because I am training a classifier which only handles **two classes**.

Specifying the thread with the parameter **"n_threads**" to speed up to process.

**Seed**, for a **random state**, to get the same results everytime.

```python
estimator_xgb = XGBClassifier(objective="binary:logistic", n_threads=4, seed=24, silent=True)
```

**Specifying the number of tunable parameters and the ranges of values for it.**

* **max_depth** = specifies the maximum depth the ensemble will go in each round.
* **n_estimator** = specifies the maximum number of rounds the ensemble will run for.
* **learning_rate** = instead of getting predictions by simply adding up the predictions from each component model, we can multiply the predictions from each model by a small number (known as the learning rate) before adding them in.

```python
parameters_XGBClass = {
    'max_depth': range (2, 10, 1),
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05]
}
```

#### Creating the GridSearch for the given parameters above:

Using the **ROC AUC metric** to compare the results of 10 fold cross validation.

```python
grid_search_XGB = GridSearchCV(
                    estimator = estimator_xgb,
                    param_grid = parameters_XGBClass,
                    scoring = 'roc_auc',
                    n_jobs = 10,
                    cv = 10,
                    verbose =0)
```

Creating a pipeline for the Grid Search:

```python
warnings.filterwarnings("ignore")


start2 = time.time()

pipeline_grid_xgb = Pipeline(steps=[('preprocessor', preprocessor),
                                ('gridsearch', grid_search_XGB)])

pipeline_grid_xgb.fit(X_train, y_train)

print(f'\n{time.time() - start2}')
```

```python
XGB_final = grid_search_XGB.best_estimator_
```

#### **Fitting the training set with the best estimator:**

```python
xgb_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                    ('model', XGB_final)])

xgb_pipeline.fit(X_train, y_train)
```

**Predicting using our XGBClassifier model.**

```python
preds_xgb = xgb_pipeline.predict(X)
```
